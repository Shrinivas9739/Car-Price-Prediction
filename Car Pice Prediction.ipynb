```python
x = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
# First apply Ohe on columns and pass the remaning columns ,,, column_tranformer is used to apply transformations of columns  (Simpleinputer removes nan values)
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
# It first applys column_trans and give it to lr
pipe = make_pipeline(column_trans,lr)
```

```python
# It directly applies Ohe on x_train columns and passes to lr
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
# To get perfect random state so our r2 score increase

for i in range(1000):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  print(r2_score(y_test,y_pred),i)
```

```python
# To check maximum random_state
score = []
for i in range(1000):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  score.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
# Now lets train model at random_state=433
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=np.argmax(score))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```
